# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('benign_0__mal_1',axis=1)
y = df['benign_0__mal_1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Create a Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
import tensorflow as tf
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,
          y=y_train,
          epochs=600,
          validation_data=(X_test,y_test),
          verbose=1)
```

```python
model_loss = model.history.history
```

```python
losses = pd.DataFrame(model_loss)
```

```python
losses.plot()
```

# Early Stopping

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1,
         callbacks=[early_stopping])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

# DropOut Layers

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1,
         callbacks=[early_stopping])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

# Model Evaluation

```python
predictions = (model.predict(X_test) > 0.5)*1
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
